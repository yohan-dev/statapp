```python
from bs4 import BeautifulSoup
import requests
import re
def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html5lib')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))
ny_bb = url_to_string('https://www.nytimes.com/2018/08/13/us/politics/peter-strzok-fired-fbi.html?hp&action=click&pgtype=Homepage&clickSource=story-heading&module=first-column-region&region=top-news&WT.nav=top-news')
article = nlp(ny_bb)
article_person = [a.text for a in article.ents if a.label_ == 'PERSON']
print(article_person)
```

```python
unique_person = []
for a in article_person:
    if a not in unique_person:
        unique_person.append(a)
final_chain = unique_person[0]
for u in unique_person[1:]:
    final_chain = final_chain + "|" + u
final_chain = final_chain.lower()
print(final_chain)
```
